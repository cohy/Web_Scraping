```python
import bs4 as bs
import request

from urllib.request import urlopen
```

```python
src =  urlopen('https://www.nytimes.com/').read()
```

```python
bsoup = bs.BeautifulSoup(src,'lxml')
```

```python
print(bsoup)
```

```python
print(bsoup.title)
```

```python
print(bsoup.title.text)
```

```python
#'a' represents anchored text
for link in bsoup.find_all('a'):
    print(link.get('href'))
    
```

```python
print(bsoup.p.text)
```

```python
#to print all the paragraph's 

print(bsoup.find_all('p'))
```

```python
#to get rid of all the tags
ptags = bsoup.find_all('p')
for p in ptags:
    print(p.text)
```

```python
src =  urlopen('http://www.espn.com/nba/statistics/player/_/stat/assists/sort/avgAssists/').read()
```

```python
bsoup = bs.BeautifulSoup(src,'lxml')
```

```python
tbl = bsoup.find('table')
```

```python
tbl_rows = tbl.find_all('tr')
for tr in tbl_rows:
    td = tr.find_all('td')
    row = [i.text for i in td]
    print(row)
```
